#### 2 .Total working days for each agents 

```python
df_perf['Working_days']=df_perf['Total Chats'].apply(lambda x:1 if x!=0 else 0)
```

```python
# Working days on weekly basis for particular Agent name taken from user

while True:
    try:
        agent=input("Enter the name of agent whose working days on weekly basis you wish to observe : ")
        if agent in list(df_perf['Agent Name']):
            print(df_perf[(df_perf['Agent Name']==agent) & (df_perf['Total Chats']!=0) ].groupby(['Agent Name','Week'])['Working_days'].sum())
            break;
        else:
            print("Enter the Agent name correctly")
    except Exception as e:
        logging.error(e)
        logging.exception(e)
        print(e)
```

#### 3. Total query that you have taken 

```python
df_perf['Month']=df_perf['Date'].dt.month
df_perf['Year']=df_perf['Date'].dt.year
```

```python
#Total query taken on weekly basis by particular Agent (agent name taken from user)

while True:
    try:
        agent=input("Enter the name of agent whose working days on weekly basis you wish to observe : ")
        if agent in list(df_perf['Agent Name']):
            print('\033[1m'+'Total query that Agent has taken on weekly basis:'+'\033[0m')
            print(df_perf[(df_perf['Agent Name']==agent)].groupby(['Agent Name','Week'])['Total Chats'].sum())
            print('\033[1m'+'Total query that Agent has taken on Monthly basis:'+'\033[0m')
            print(df_perf[(df_perf['Agent Name']==agent)].groupby(['Agent Name','Month'])['Total Chats'].sum())
            print('\033[1m'+'Total query that Agent has taken on Yearly basis:'+'\033[0m')
            print(df_perf[(df_perf['Agent Name']==agent)].groupby(['Agent Name','Year'])['Total Chats'].sum())
            
            break;
        else:
            print("Enter the Agent name correctly")
    except Exception as e:
        logging.error(e)
        logging.exception(e)
        print(e)
```

#### 4. total Feedback that you have received 

```python
#Total Feedback received for particular Agent name taken from user

while True:
    try:
        agent=input("Enter the name of agent whose working days on weekly basis you wish to observe : ")
        if agent in list(df_perf['Agent Name']):
            print('\033[1m'+'Total Feedback that Agent has received on weekly basis:'+'\033[0m')
            print(df_perf[(df_perf['Agent Name']==agent)].groupby(['Agent Name','Week'])['Total Feedback'].sum())
            print('\033[1m'+'Total Feedback that Agent has received on Monthly basis:'+'\033[0m')
            print(df_perf[(df_perf['Agent Name']==agent)].groupby(['Agent Name','Month'])['Total Feedback'].sum())
            print('\033[1m'+'Total Feedback that Agent has received on Yearly basis:'+'\033[0m')
            print(df_perf[(df_perf['Agent Name']==agent)].groupby(['Agent Name','Year'])['Total Feedback'].sum())
            
            break;
        else:
            print("Enter the Agent name correctly")
    except Exception as e:
        logging.error(e)
        logging.exception(e)
        print(e)
```
