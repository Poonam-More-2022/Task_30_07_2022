```python
def total_hours(l1):
    ''' This function returns total hours on weekly basis for a particular agent '''
    try: 
        if type(l1)==list:
            tsum=dt.timedelta()
            for i in l1:
                #print(i)
                if type(i)==dt.time:
                    h,m,s=str(i).split(':')
                    #print(h,m,s)
                    tsum=tsum+dt.timedelta(hours=int(h),minutes=int(m),seconds=int(s))
                    #print(respsum)  
                elif type(i)==dt.datetime:
                    tme = datetime.strptime(str(i),'%Y-%m-%d %H:%M:%S').time()
                    h,m,s=str(tme).split(':')
                    tsum=tsum+dt.timedelta(hours=int(h),minutes=int(m),seconds=int(s))
                else:
                    pass
            logging.info(" The total contribution hours on weekly basis for a given agent is %s",tsum)
            return tsum
        else:
            pass
    except Exception as e:
        logging.error(e)
        logging.exception(e)
        return e
```

```python
def total_hours_per(l1):
    ''' This function returns total % hours on Monthly basis for a particular agent '''
    try: 
        if type(l1)==list:
            tsum=dt.timedelta()
            for i in l1:
                #print(i)
                if type(i)==dt.time:
                    h,m,s=str(i).split(':')
                    #print(h,m,s)
                    tsum=tsum+dt.timedelta(hours=int(h),minutes=int(m),seconds=int(s))
                    #print(respsum)  
                elif type(i)==dt.datetime:
                    tme = datetime.strptime(str(i),'%Y-%m-%d %H:%M:%S').time()
                    h,m,s=str(tme).split(':')
                    tsum=tsum+dt.timedelta(hours=int(h),minutes=int(m),seconds=int(s))
                else:
                    pass
            logging.info(" The total contribution hours on Monthly basis for a given agent is %s",tsum)
            return (tsum/(dt.timedelta(hours=int(24*5*4),minutes=int(0),seconds=int(0))))*100
        else:
            pass
    except Exception as e:
        logging.error(e)
        logging.exception(e)
        return e
```

```python
# total percentage of active hours on Monthly basis for each agent

while True:
    try:
        agent_names=list(df_log['Agent'].unique())
        data=pd.DataFrame()
        Monthly_duration={}
        for name in agent_names:
            b=list(df_log[df_log['Agent']==name].groupby(['Agent','Month'])['Duration'])
            for i in range(len(b)):
                month_no=b[i][0][1]
                dur_list=list(b[i][1])
                dur=total_hours_per(dur_list)
                Monthly_duration['Agent Name']=name
                Monthly_duration['Month']=month_no
                Monthly_duration['Duration']=str(dur)
                data=data.append(Monthly_duration, ignore_index=True)  
        print(data.groupby(['Agent Name', 'Month'])['Duration'].first())
        break;
        
    except Exception as e:
        logging.error(e)
        logging.exception(e)
        print(e)
```
